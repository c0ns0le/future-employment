# WORK ACTIVITIES

```python
dwa_ref = pd.read_csv('data/databases/db15/DWA Reference.txt', delimiter='\t')
iwa_ref = pd.read_csv('data/databases/db15/IWA Reference.txt', delimiter='\t')
dwas = pd.read_csv('data/databases/db15/Tasks to DWAs.txt', delimiter='\t')

def get_n_most_popular(n, IWA = False, print_tasks = False):
    new_dwas = dwas.merge(dwa_ref, on = 'DWA ID', how = 'inner')
    if IWA:
        group_var = 'IWA ID'
        join_df = iwa_ref
    else:
        group_var = 'DWA ID'
        join_df = dwa_ref
    gb = new_dwas.groupby(group_var).count()
    new_gb = gb.sort_values(by = 'Date', ascending = False)
    subset = new_gb.iloc[:n,:].reset_index()[[group_var, 'Task ID']]
    subset.columns = [group_var, 'count']
    subset = subset.merge(join_df, on = group_var, how = 'inner')
    if print_tasks:
        col_title = group_var[:3] + " Title"
        for i in subset[col_title]:
            print i, '\n'
    else:
        return subset
    
# dwa_ref[dwa_ref['DWA ID'].isin(ten_most_popular)]
```

```python
top50 = get_n_most_popular(50, True)
nums = top50['count']
title = top50['IWA Title']

print "id", "| # of occupations with this activity "
print "----------------------------------"
for i, num in enumerate(nums):
    print i, " | ", num, " | ", title[i]
```

# SPEECH RECOGNITION

```python
import speech_recognition
import pyttsx
```

```python
speech_engine = pyttsx.Engine() # see http://pyttsx.readthedocs.org/en/latest/engine.html#pyttsx.init
speech_engine.setProperty('rate', 150)
```

```python
def speak(text):
	speech_engine.say(text)
	speech_engine.runAndWait()

recognizer = speech_recognition.Recognizer()

def listen():
	with speech_recognition.Microphone() as source:
		recognizer.adjust_for_ambient_noise(source)
		audio = recognizer.listen(source)

	try:
		return recognizer.recognize_sphinx(audio)
		# or: return recognizer.recognize_google(audio)
	except speech_recognition.UnknownValueError:
		print("Could not understand audio")
	except speech_recognition.RequestError as e:
		print("Recog Error; {0}".format(e))

	return ""

speak("Say something!")
speak("I heard you say " + listen())
```
