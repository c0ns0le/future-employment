```python
def get_name_codes():
    data = pd.read_table('../data/onet_tools_technology/Tools and Technology.txt')
    df = data.drop(['T2 Type', 'T2 Example', 'Commodity Code', 'Commodity Title'], axis = 1).groupby(['O*NET-SOC Title', 'O*NET-SOC Code']).head()
    gb = df.groupby('O*NET-SOC Title')
    names = gb.head().drop_duplicates()
    return names.reset_index(drop = True)
```
