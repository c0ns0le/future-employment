```python
def get_percentiles(array):
	def percentile(x, array):
		return 100*np.mean(array <= x)

	return np.array(map(lambda x: percentile(x, array), array))
```

```python
def clean_element(element):
        try:
            e = str(element)
            e = e.replace(',', "").replace('$','')
            return float(e)
        except ValueError:
            return None
```

```python
def clean_number_array(array):
    def clean_element(element):
        try:
            e = str(element)
            e = e.replace(',', "").replace('$','')
            return float(e)
        except ValueError:
            return None
    return map(lambda x: clean_element(x), array)

def clean_question_array(array):
    def clean_element(element):
        if element != '\xd1':
            return element
        else:
            return None
        
    return map(lambda x: clean_element(x), array)
```

```python
stats.iloc[:,2:-3] = stats.iloc[:,2:-3].apply(clean_number_array, axis = 1)
stats.iloc[:,-3:] = stats.iloc[:,-3:].apply(clean_question_array, axis = 1)
```
